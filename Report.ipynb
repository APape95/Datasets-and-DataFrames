```python
import pandas as pd
```

```python
df = pd.read_csv('balance.txt', delim_whitespace=True)

df.head()
```

```python
# Compare the average income based on ethnicity.
print(f"The average income based on ethnicity is: {df.groupby('Ethnicity')['Income'].mean()}")
```

```python
# On average, do married or single people have a higher balance?
# Group by Married to show results based on the result of this coumn
# Showing the mean of the balance per catogory
print(df.groupby('Married')['Balance'].mean())
```

```python
# What is the highest income in our dataset?
print("The highest income in our dataset is", df.loc[:, "Income"].max())
```

```python
# What is the lowest income in our dataset?
print("The lowest income in our dataset is", df.loc[:, "Income"].min())
```

```python
# How many cards do we have recorded in our dataset? (Hint: use `sum()`)
cards_ttl = df.loc[:,"Cards"].sum()
print(f"We have a total of {cards_ttl} cards within our data set")
```

```python
# How many females do we have information for vs how many males? 
# (Hint: use `count()` for a list of all methods for computation of descriptive stats, explore the 
# [pandas documentation](https://pandas.pydata.org/pandas-docs/stable/reference/frame.html#computations-descriptive-stats)).

female_count = df["Gender"].value_counts()["Female"]
male_count = df["Gender"].value_counts()["Male"]

print(f"We have {female_count} Females within this data compared to {male_count} Males")
```
